```python
import re
import os
import sys
import nltk
import spacy
import PyPDF2
from nltk.corpus import stopwords
from nltk.cluster.util import cosine_distance

def read_pdf(filename):
    # Open the PDF file
    file = open(filename, 'rb')

    # Create a PDF object
    pdf = PyPDF2.PdfReader(file)

    # Get the number of pages in the PDF
    num_pages =len(pdf.pages)

    # Initialize a list to store the text of each page
    text = []

    # Loop through all the pages and extract the text
    for i in range(num_pages):
        page = pdf.pages[i]
        
        text.append(page.extract_text())

    # Close the file
    file.close()

    # Return the text as a single string
    return '\n'.join(text)

def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Tokenize the text
    tokens = nltk.word_tokenize(text)

    # Remove stopwords
    stop_words = stopwords.words('english')
    tokens = [token for token in tokens if token not in stop_words]

    # Use spacy to lemmatize the tokens
    nlp = spacy.load("en_core_web_sm")
    tokens = [token.lemma_ for token in nlp(str(tokens))]

    return tokens

def get_similarity(tokens1, tokens2):
    # Get the similarity score between the two sets of tokens
    return nltk.cluster.util.cosine_distance(tokens1, tokens2)

def get_summary(text, num_sentences):
    # Preprocess the text
    tokens = preprocess_text(text)

    # Split the text into sentences
    sentences = nltk.sent_tokenize(text)

    # Initialize a list to store the summary
    summary = []

    # Loop through all the sentences
    for sentence in sentences:
        # Get the tokens for the sentence
        sentence_tokens = preprocess_text(sentence)

        # Calculate the similarity between the sentence tokens and the overall text tokens
        similarity=0.6
        try:
            similarity = get_similarity(tokens, sentence_tokens)
        except:
            print("something")

        # If the similarity is high, add the sentence to the summary
        if similarity < 0.5:
            summary.append(sentence)

        # If the summary is long enough, stop
        if len(summary) >= num_sentences:
            break

    return summary

def main():
    # Read the PDF file
    text = read_pdf('test_paper.pdf')

    # Get the summary
    summary = get_summary(text, 3)

    # Print the summary
    for sentence in summary:
        print(sentence)

if __name__ == '__main__':
    main()
```
